# Report: Analysis of Particulate matter (PM 2.5) and (PM 10) in Nigeria(2020)

This Analysis involves the calculation of PM 2.5 and PM 10 in Nigeria for the year 2020.

## Data Used:
1. The data used was obtained from ads copernicus with spatial resolution 0.1x0.1 degree in Nigeria for 2020.
2. It was downloded in 3hourly 

## Methodology
### 1.Diurnal
The diurnal data was obtained by averaging all the values for each hour of the days: 0:00, 3:00, 6:00, 9:00, 12:00, 15:00, 18:00, and 21:00, for the entire year to just one value for these hours. This process resulted in a single value for each hour

### 1. Monthly
The monthly data was extracted using Microsoft Excel. Each month's data was calculated by averaging all the values for each hour. For example, in January, the values for 0:00, 3:00, 6:00, 9:00, 15:00, 18:00, and 21:00 were averaged to obtain a single value for each hour of the day. This process was repeated for every month.

The files were individually extracted using Microsoft Excel and then grouped accordingly using this formula:

                    Monthly= df.groupby("time").mean
                                
### 2. Seasonal
DJF,MAM,SON,JJA
The average of DJF was obtained by summing all the values for these three months and finding their average.This process was repeated for every season.

## Result
Diurnal observation: Two peaks (an increase in PM2.5 and PM10) were observed during peak morning hours (3:00 and 6:00 Hrs) and late evening (12:00 and 15:00Hrs). However, the percentage reduction remains almost constant throughout the day.

Seasonal observation: We experience the highest PM concentration (2.5 and 10) in DJF, the dry season and the lowest in JJA,the wet season.

Monthly observation: We experience the highest PM concentration (2.5 and 10) in February,dry season and the lowest in July, wet season.

Overall, the relationship between PM2.5 and PM10 is not always linear some are simple, while others are complex because of the complex chemistry and atmospheric dynamic
